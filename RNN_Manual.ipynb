```
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
%matplotlib inline
```

```
inputs_num=2
```

```
num_neurons=3
```

```
#placeholders
x0= tf.placeholder(tf.float32,[None,inputs_num])
```

```
x1= tf.placeholder(tf.float32,[None,inputs_num])
```

```
#variables
wx=tf.Variable(tf.random_normal(shape=[inputs_num,num_neurons]))
```

```
wy=tf.Variable(tf.random_normal(shape=[num_neurons,num_neurons]))
```

```
b=tf.Variable(tf.zeros(shape=[1,num_neurons]))
```

```
#graphs
y0=tf.tanh(tf.matmul(x0,wx)+b)
y1=tf.tanh(tf.matmul(y0,wy)+ tf.matmul(x1,wx)+b)
```

```
init=tf.global_variables_initializer()
```

```
#creating data
x0_batch=np.array([[0,1],[2,3],[4,5]]) #at timestamp 0

x1_batch=np.array([[100,101],[102,103],[104,105]]) #at timestamp 1

```

```
with tf.Session() as sess:
  sess.run(init)

  y0_out,y1_out=sess.run([y0,y1],feed_dict={x0:x0_batch,x1:x1_batch})
```
